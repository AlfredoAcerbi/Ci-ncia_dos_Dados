## Q2 - Fit e Histograma

Faça o Histograma da Magnitude. Interprete.

```python
faixas = np.arange(df['Magnitude'].min(), df['Magnitude'].max(), 0.25)
plot=df['Magnitude'].plot.hist(bins=faixas, title='Histograma Magnitude', figsize=(6,6), alpha=0.5)
plt.xlabel('Magnitude')
```

**Quanto maior a magnitude, menor a frequencia. A faixa de maior frequencia é de 5.5 a 6.**

Faça o fit de uma distribuição exponencial sobre os dados da Magnitude, achando os valores de **loc** e **scale**. Interprete loc e scale no caso da exponencial.
Documentação: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html

```python
mu = df['Magnitude'].mean()
lamb = 1/mu
sig = 1/lamb
n = 100
scale = mu
loc = 0

mag = expon.rvs(loc,scale,n)
```

Refaça o Histograma plotando a fdp (função densidade de probabilidade) da exponencial com os parâmetros achados no fit em cima. Cuidado com o domínio utilizado. Interprete.

```python
faixa2 = np.arange(mag.min(),mag.max(), 0.1)

plt.hist(mag, normed = True)
plt.plot(faixa2, expon.pdf(faixa2, loc, scale))
plt.title('Histograma e FDP')
plt.legend(['FDP', 'Histograma'])
```

### Q3.1 - Qual a probabilidade de ocorrer um terremoto no hemisfério norte?

```python
tc = pd.crosstab(df['Hemisfério'], df['Type'], margins=True,rownames=['Hemisfério'], colnames=['Type'], normalize=True)
tc
```

```python
prob_terr_norte = 0.447140 * 100

print ("A probabilidade de ocorrer um terremoto no hemisfério norte é de {0} %" .format(prob_terr_norte))
```

## Q4 - Análise bivariada

Faça o *plot* de dispersão (*scatter plot*) entre as variáveis `Magnitude Error`  e `Depth`

```python
df2 = df.dropna(subset=['Magnitude Error', 'Depth'])

analise = plt.scatter(df2['Magnitude Error'], df2['Depth'])
plt.show()
```

Calcule a correlação entre as variáveis `Magnitude Error`  e `Depth`

```python
corr = np.corrcoef(df2['Magnitude Error'],df2.Depth)[0,1]
```

```python
print ("A correlação entre as variáveis Magnitude Error e Depth é {0}" .format(corr))
```
